# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/AshishJangra27/Skill-India-AI-ML-Scholarship/main/Summer%20Olympics/summer.csv")
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
df.groupby(['Medal'])
medal_grp = df.groupby(['Medal'])
medal_grp.get_group("Gold")
```

```python
medal_grp['Sport'].value_counts().loc["Gold"].head()
gold = medal_grp['Sport'].value_counts().loc["Gold"].head()
```

```python
gold
```

```python
plt.figure(figsize=(5,5))
plt.title('sports having max number of  gold medals')
sns.barplot(x=gold.index,y=gold);
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
sport = df['Sport'].value_counts().head()
sport
```

```python
plt.figure(figsize=(5,5))
plt.title('Top 5 sports having max number of medals')
sns.barplot(x=sport.index,y=sport);
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
year = df[(df.Country=='IND') & (df.Medal=='Gold')].head(15)
year
```

```python
first_gold = year['Year'].head(15)
first_gold
```

```python
plt.figure(figsize=(10,5))
plt.title('In 1928 INDIA won first gold medal')
sns.barplot(x=first_gold.index,y=first_gold);
```
